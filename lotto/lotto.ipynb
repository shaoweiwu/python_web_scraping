```python
import pandas as pd
import numpy as np
from selenium.webdriver.edge.service import Service as EdgeService
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from time import sleep
```

```python
edge_driver_path = "msedgedriver.exe"
# edge_options = Options()
# edge_options.set_capability("args",["inprivate", "disable-notifications"])

# driver = webdriver.Edge(executable_path=edge_driver_path, capabilities=edge_options)
driver = webdriver.Edge(executable_path=edge_driver_path)

url = 'url'
driver.get(url)
```

```python
df = pd.read_csv('lotto_history.csv', encoding='utf-8')
df
```

```python
# 更新樂透資料用

# 找出以期別作為選擇的按鈕並點擊
lotto_issue_button = driver.find_element(By.ID, "Lotto649Control_history_radNO")
lotto_issue_button.click()

last_issue_number = df['期別'].astype(int).max()+1    
while True:
    sleep(1)
    count=0
    
    # 找到要填入期別的格子並逐一填入最大期別+1的數值來更新
    lotto_issue_number = driver.find_element(By.ID, "Lotto649Control_history_txtNO")
    lotto_issue_number.clear()
    lotto_issue_number.send_keys( f"{last_issue_number}" )
    
    search_button = driver.find_element(By.ID, "Lotto649Control_history_btnSubmit")
    search_button.click()
    
    last_one = driver.find_element(By.ID, "Lotto649Control_history_Label1").text
    if last_one=='查無資料':
        last_issue_number = int((last_issue_number+1000000)/1000)*1000+1
        lotto_issue_number = driver.find_element(By.ID, "Lotto649Control_history_txtNO")
        lotto_issue_number.clear()
        lotto_issue_number.send_keys( f"{last_issue_number}" )
        search_button = driver.find_element(By.ID, "Lotto649Control_history_btnSubmit")
        search_button.click()
        
        last_one = driver.find_element(By.ID, "Lotto649Control_history_Label1").text
        if last_one!='查無資料':
            continue
        else:
            break
    
    # 防呆
    if count==110:
        break
    count+=1
    
    
    data = []
    # 期別
    issue = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_DrawTerm_0']")
    # 開獎日
    date = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_DDate_0]")
    # 銷售金額
    sellAmount = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_SellAmount_0]")
    # 獎金總額
    totalPrize = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_Total_0]")
    # 開獎號碼
    SNo1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo1_0]")
    SNo2 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo2_0]")
    SNo3 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo3_0]")
    SNo4 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo4_0]")
    SNo5 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo5_0]")
    SNo6 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo6_0]")
    SNo7 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_No7_0]")
    numOfPrize1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA3_0]")
    prize1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA4_0]")
    accumulated1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA5_0]")

    data = [issue.text, date.text, sellAmount.text, totalPrize.text,
            SNo1.text, SNo2.text, SNo3.text, SNo4.text, SNo5.text, SNo6.text, SNo7.text,
            numOfPrize1.text, prize1.text, accumulated1.text]
    for count in range(0, 17):
        if count==0:
            data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_CategB3_0']")
            data.append(data1.text)

        elif count==1:
            data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_CategC3_0']")
            data.append(data1.text)

        else:
            data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_Label{count}_0']")
            data.append(data1.text)
    
    new_data = pd.DataFrame([data], columns=df.columns)
    df = pd.concat([df, new_data], ignore_index=True)
    last_issue_number+=1
    
df

# df.to_csv('lotto_history.csv', encoding='utf-8-sig', index=False)

# 關閉瀏覽器
driver.quit()
```

```python
# 抓樂透全部歷史資料用

%%time
# 找出以下拉式選單作為選擇的按鈕並點擊
lotto_issue_button = driver.find_element(By.ID, "Lotto649Control_history_radYM")
lotto_issue_button.click()

# 找到'年份'的下拉式選單並作為select的對象
lotto_issue_year_dropdown = driver.find_element(By.ID, "Lotto649Control_history_dropYear")
lotto_issue_year_dropdown = Select(lotto_issue_year_dropdown)

# 找到'月份'的下拉式選單並作為select的對象
lotto_issue_month_dropdown = driver.find_element(By.ID, "Lotto649Control_history_dropMonth")
lotto_issue_month_dropdown = Select(lotto_issue_month_dropdown)


df = pd.DataFrame()
for year in range(0, len(lotto_issue_year_dropdown.options)):
    lotto_issue_year_dropdown.select_by_index(year)
    for month in range(0, len(lotto_issue_month_dropdown.options)):
        lotto_issue_month_dropdown.select_by_index(month)
        #找到'查詢'格子並點擊
        search_button = driver.find_element(By.ID, "Lotto649Control_history_btnSubmit")
        search_button.click()
        
        # 當label1文字為'查無資料'則跳出迴圈
        last_one = driver.find_element(By.ID, "Lotto649Control_history_Label1").text
        if last_one=='查無資料':
            break
        sleep(3)
        
        # 找出當月所有期數
        issueL = driver.find_elements(By.CSS_SELECTOR, "span[id^=Lotto649Control_history_dlQuery_L649_DrawTerm_]")

        for issue_index in range(0, len(issueL)):
            data = []

            # 期別
            issue = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_DrawTerm_{issue_index}']")
            # 開獎日
            date = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_DDate_{issue_index}]")
            # 銷售金額
            sellAmount = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_SellAmount_{issue_index}]")
            # 獎金總額
            totalPrize = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_Total_{issue_index}]")
            # 開獎號碼
            SNo1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo1_{issue_index}]")
            SNo2 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo2_{issue_index}]")
            SNo3 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo3_{issue_index}]")
            SNo4 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo4_{issue_index}]")
            SNo5 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo5_{issue_index}]")
            SNo6 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_SNo6_{issue_index}]")
            SNo7 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_No7_{issue_index}]")
            numOfPrize1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA3_{issue_index}]")
            prize1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA4_{issue_index}]")
            accumulated1 = driver.find_element(By.CSS_SELECTOR, f"span[id=Lotto649Control_history_dlQuery_L649_CategA5_{issue_index}]")

            data = [issue.text, date.text, sellAmount.text, totalPrize.text,
                    SNo1.text, SNo2.text, SNo3.text, SNo4.text, SNo5.text, SNo6.text, SNo7.text,
                    numOfPrize1.text, prize1.text, accumulated1.text]
            
            # 中獎注數、中獎金額、累計金額
            # 橘色格子(除了貳獎和參獎的中獎注數以外，其他的都是用label，且一橘一綠搭配共34個)
            if issue_index%2==0:
                for count in range(0, 17):
                    if count==0:
                        data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_CategB3_{issue_index}']")
                        data.append(data1.text)

                    elif count==1:
                        data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_L649_CategC3_{issue_index}']")
                        data.append(data1.text)

                    else:
                        data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_Label{count}_{issue_index}']")
                        data.append(data1.text)
            # 綠色格子(全都是label)
            else:
                for count in range(18, 35):
                    data1 = driver.find_element(By.CSS_SELECTOR, f"span[id='Lotto649Control_history_dlQuery_Label{count}_{issue_index}']")
                    data.append(data1.text)
            df = df.append([data])        
        
        lotto_issue_year_dropdown = driver.find_element(By.ID, "Lotto649Control_history_dropYear")
        lotto_issue_year_dropdown = Select(lotto_issue_year_dropdown)

        lotto_issue_month_dropdown = driver.find_element(By.ID, "Lotto649Control_history_dropMonth")
        lotto_issue_month_dropdown = Select(lotto_issue_month_dropdown)
        print(f"{lotto_issue_year_dropdown.options[year].text}年{lotto_issue_month_dropdown.options[month].text}月 done")
        
column_names = ["期別", "開獎日", "銷售金額", "獎金總額", "獎號1",
                "獎號2", "獎號3", "獎號4", "獎號5", "獎號6", "特別號",
                "頭獎注數", "頭獎獎金", "頭獎累至次期獎金", 
                "貳獎注數", "參獎注數", "肆獎注數", "伍獎注數","陸獎注數", "柒獎注數", "普獎注數",
                "貳獎獎金", "參獎獎金", "肆獎獎金", "伍獎獎金", "陸獎獎金", "柒獎獎金", "普獎獎金",
                "貳獎累至次期獎金", "參獎累至次期獎金", "肆獎累至次期獎金"
                ]
df.columns = column_names

# df.to_csv('lotto_history.csv', encoding='utf-8-sig', index=False)

# 關閉瀏覽器
driver.quit()
```
