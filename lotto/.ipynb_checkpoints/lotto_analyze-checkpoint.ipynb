```python
df = pd.read_csv('lotto_analyze.csv', encoding='utf-8')
df
```

```python
SNo_df = pd.concat([df['獎號1'], df['獎號2'], df['獎號3'], df['獎號4'], df['獎號5'], df['獎號6'], df['特別號']], axis=1)
SNo_df
```

```python
from collections import Counter
```

```python
SNo_df.values
```

```python
SNo_df.values.tolist()
```

```python
SNo_stack = SNo_df.stack()
SNo_stack
```

```python
SNo_stack.values
```

```python
SNo_stack.tolist()
```

```python
counter = Counter(SNo_stack.tolist())
counter
```

```python
SNos = list(counter.keys())
SNos_count = list(counter.values())

plt.figure(figsize=(10, 6))
plt.bar(SNos, SNos_count)
plt.xlabel('numbers')
plt.ylabel('times')
plt.title('lotto history count of numbers')
plt.show()
```

```python
SNo_df.columns
```

```python
for i in SNo_df.columns:
    SNo = Counter(SNo_df[i].tolist())
    SNos = list(SNo.keys())
    SNo_count = list(SNo.values())

    plt.figure(figsize=(10, 6))
    plt.bar(SNos, SNo_count)
    plt.xlabel('numbers')
    plt.ylabel('times')
    plt.title(f'lotto history count of SNo{i}')
    plt.show()
```

```python
SNo_stack16 = SNo_df[['獎號1', '獎號2', '獎號3', '獎號4', '獎號5', '獎號6']].stack()
SNo = Counter(SNo_stack16.tolist())
SNos = list(SNo.keys())
SNo_count = list(SNo.values())

plt.figure(figsize=(10, 6))
plt.bar(SNos, SNo_count)
plt.xlabel('numbers')
plt.ylabel('times')
plt.title('lotto history count of SNo7')
plt.show()
```

```python
SNo7 = Counter(SNo_df['特別號'].tolist())
SNos = list(SNo7.keys())
SNo7_count = list(SNo7.values())

plt.figure(figsize=(10, 6))
plt.bar(SNos, SNo7_count)
plt.xlabel('numbers')
plt.ylabel('times')
plt.title('lotto history count of SNo7')
plt.show()
```
